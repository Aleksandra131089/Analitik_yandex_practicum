## Предобработка данных

### Обработка пропусков

```python
#убираем строки с пропусками в locality_name, floors_total
data.dropna(subset=["locality_name", "floors_total"], inplace=True)
```

Так как количество пропусков в этих столбцах мало, удалили их, чтобы в дальнейшем они не мешали нам в расчетах.

```python
#заполняем пропуски в жилой площади медианным значением в зависимости от количества комнат
data["living_area"] = data["living_area"].fillna(data.groupby("rooms")["living_area"].transform("median"))
display(data["living_area"].isna().sum())
```

```python
#заменим букву ё на е в населенных пунктах и исправим дубликаты
data["locality_name"] = data["locality_name"].str.replace("ё", "е")
data["locality_name"] = data["locality_name"].where(data["locality_name"] != "Кудрово", "деревня Кудрово")
data["locality_name"] = data["locality_name"].where(data["locality_name"] != "Мурино", "поселок Мурино")
data["locality_name"] = data["locality_name"].where(data["locality_name"] != "городской поселок Большая Ижора", "поселок городского типа Большая Ижора")
data["locality_name"] = data["locality_name"].where(data["locality_name"] != "городской поселок Рощино", "поселок городского типа Рощино")
data["locality_name"] = data["locality_name"].where(data["locality_name"] != "городской поселок Мга", "поселок городского типа Мга")
data["locality_name"] = data["locality_name"].where(data["locality_name"] != "городской поселок Назия", "поселок городского типа Назия")
display(data["locality_name"].value_counts())
```

<div class="alert alert-info">
В целом заменяла те названия, которые бросились в глаза, т.к. они мало влияют на структуру датафрейма решила ограничится только этими шестью. </div>

```python
#меняем количество балконов с Nan
data["balcony"] = data["balcony"].fillna(0)
display(data["balcony"].isna().sum())
```

Исходя из логики. что в домах одинаковой этажности потолки должны быть приблизительно одной высоты, заменим пропуски медианными значениями в зависимости от этажности дома.

```python
#заполняем высоту потолков
data["ceiling_height"] = data["ceiling_height"].fillna(data.groupby("floors_total")["ceiling_height"].transform("median"))
display(data["ceiling_height"].isna().sum())
```

```python
#переведем цену квартир в тыс.кв.м, чтобы облегчить расчеты.
data["last_price"] = data["last_price"] / 1000
```

Постараемся заполнить пропуска в площади кухонь опираясь на общую и жилую площадь

```python
#посчитаем разницу между общей и жилой площадью
data["area"] = data["total_area"] - data["living_area"]
#заполним пропуска в площади кухонь
data["kitchen_area"] = data["kitchen_area"].fillna(data.groupby(["rooms", "total_area"])["area"].transform("median"))
display(data["kitchen_area"].isna().sum())
```

### Замена типов данных

```python
#меняем тип данных с float на int

data["balcony"] = data["balcony"].astype("int")
data["floors_total"] = data["floors_total"].astype("int")
data["last_price"] = data["last_price"].astype("int")
```

```python
#переведем тип данных в столбце "first_day_exposition" в datetime
data["first_day_exposition"] = pd.to_datetime(data["first_day_exposition"], format="%Y-%m-%dT%H:%M:%S")
display(data["first_day_exposition"].head())
```

```python
# приведем колонку is_apartment к булевому типу
data["is_apartment"] = data["is_apartment"].astype("bool")
```

```python
data = data.reset_index(drop=True)
#посмотрим как изменилась общая структура нашего датафрейма после предобработки
data.info()
```

### Вывод 
   
   Ознакомившись с данными мы выявили ряд проблем, от которых постарались избавиться на этапе предобработки. В частности мы удалили пропуски, там, где их количество некритично, заменили медианными значениями пропуски в тех столбцах где можно было опереться на другие столбцы. 
        
   Также перевели некоторые данные в удобоваримый формат с `float` на `int`, и с `object` на `datetime`. Поработали со списком населенных пунктов убрав неявные дубликаты и изменив букву ё на е. После окончания предобработки мы избавились всего от **173** строк датафрейма, что не должно сказаться на качестве наших данных. К сожалению, не удалось найти адекватной замене пропускам в `"cityCenters_nearest"` и `"days_exposition"`, но их много, удалить без потери точности не удастся, поэтому эти столбцы оставим без изменений. 
   
   Остальные столбцы с пропусками как то наличие парков, прудов или удаленность от аэропорта не нужны нам для дальнейшего исследования, поэтому оставим их как есть.

## Расчёты и добавление результатов в таблицу

Посчитаем и добавим в таблицу основные показатели необходимые нам для дальнейшего анализа:
- цену квадратного метра;
- день недели, месяц, год публикации объявления;
- этаж квартиры(первый, последний, другой);
- соотношение жилой и общей площади;
- отношение площади кухни к общей.

### Расчет цены квадратного метра

```python
#расчитываем цену квадратного метра жилья
data["meter_price"] = round((data["last_price"] / data["total_area"]), 2)
display(data["meter_price"].head(10))
display(data["meter_price"].describe())
```

Медианое значение - **95тыс.руб**. Среднее - **99тыс.руб**

### Добавление дня, месяца, года

```python
#создаем столбцы с днем недели, месяцем, годом публикации
data["day_of_week"] = data["first_day_exposition"].dt.weekday
data["month"] = data["first_day_exposition"].dt.month
data["year"] = data["first_day_exposition"].dt.year
data['day_name_rev'] = data['first_day_exposition'].dt.day_name()
display(data['day_name_rev'].unique())
data.columns
```

### Категоризация этажей

```python
#определяем категорию этажа 
def floor_category (row):
    if row["floor"] == 1:
        return "первый"
    if row["floor"] == row["floors_total"]:
        return "последний"
    if (row["floor"] != 1) and (row["floor"] != row["floors_total"]):
        return "другой"
    
data["floor_category"] = data.apply(floor_category, axis=1)
display(data["floor_category"].head(15))
```

### Расчет соотношения площадей

```python
#считаем соотношение жилой и общей площади
data["percent_living_area"] = round((data["living_area"] / data["total_area"]) * 100, 2)
display(data["percent_living_area"].head(10))
```

```python
#считаем отношение площади кухни к общей
data["percent_kitchen_area"] = round((data["kitchen_area"] / data["total_area"]) * 100, 2)
display(data["percent_kitchen_area"].head(10))
```

```python
display(data.head(10))
```

### Вывод

Мы добавили в таблицу столбцы с расчетами отношения жилой и общей площади, отношения площади кухни к общей, и столбцы с днем, месяцем и годом продажи квартиры, категоризировали этажи и готовы к исследованию данных.

- Взаимосвязь цены прослеживается с площадью квартиры, слабая корреляция с количеством комнат и совсем не зависит от удаленности от центра, что странно, т.к.ранее мы смотрели на стоимость квадратного метра и видели явную зависимость от расстояния. Высота потолков не влияет на стоимость.
- В Петербурге тенденция сохраняется: первый этаж - самый дешевый, но вот последние этажи не так сильно разнятся по стоимости от других. Видимо из-за того что из окон в историческом центре открываются прекрасные виды и возможно из-за доступа к крышам.
- По субботам размещают самые дешевые варианты. Хотя так же видно проседание в летние месяца как и по общему датасету, но минимум наблюдается в декабре
- Как и в случае с общим датафреймом тенденция в Санкт-Петербурге показывает о спаде стоимости жилья начиная с **2014года**. Но в отличие от общих данных возврат к повышению стоимости идет очень медленно.
